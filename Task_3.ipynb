# To Explore Decision Tree Algorithm
Decision tree is a flowchart like tree structure, where each internal node denotes a test on an attribute, each branch represents an outcome of the test, and each leaf node (terminal node) holds a class label.

### Task :- For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.


```python
# Import library 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
# load data 
iris = datasets.load_iris()

# Creating iris dataframe
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
# Create target variable y
y = iris.target
print(y)
```

# Import Decision tree :-

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
# fir the data 
clf.fit(df,y)
```

# Visualize the Decision Tree :-

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(25,20))

tree_image = plot_tree(clf,feature_names= iris.feature_names ,class_names=iris.target_names,
                       precision=4,label="all",filled=True,rounded=True)
plt.savefig('showTree')
```

### We can now feed any new/test data to this classifer and it would be able to predict the right class accordingly.
